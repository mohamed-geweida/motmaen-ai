#Prepare Data

```python
train_path = "/content/food/food photos/train/images"
test_path = "/content/food/food photos/test/images"
valid_path = "/content/food/food photos/valid/images"
```

- To see if images shapes are equals

```python
def check_image_shapes(folder_path, num_to_show=5):
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.png', '.jpeg'))]

    print(f"Found {len(image_files)} images in {folder_path}")

    for i, image_file in enumerate(image_files[:num_to_show]):
        img_path = os.path.join(folder_path, image_file)
        img = cv2.imread(img_path)
        if img is not None:
            h, w, c = img.shape
            print(f"{image_file}: {w}x{h}, Channels: {c}")
        else:
            print(f"Could not read {image_file}")
```

```python
print("Train images:")
check_image_shapes(train_path)
print("\nTest images:")
check_image_shapes(test_path)
print("\nvalid images:")
check_image_shapes(test_path)
```

- need to fix images shapes

- Using padding to save image in the same coordiantes and coordiantes of boundry box

```python
def resize_and_pad(image_path):
    img = cv2.imread(image_path)
    h, w = img.shape[:2]
    target_h, target_w = 416,416
    # Percentage to bigger one
    scale = min(target_w / w, target_h / h)
    new_w = int(w * scale)
    new_h = int(h * scale)
    resized_img = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_AREA)
    # Padding
    delta_w = target_w - new_w
    delta_h = target_h - new_h
    top, bottom = delta_h // 2, delta_h - (delta_h // 2)
    left, right = delta_w // 2, delta_w - (delta_w // 2)
    padded_img = cv2.copyMakeBorder(resized_img, top, bottom, left, right,
                                     cv2.BORDER_CONSTANT, value=[0, 0, 0]) # Pad with black
    cv2.imwrite(image_path, padded_img)
    # return padded_img
```

```python
for filename in os.listdir(train_path):
    full_path = os.path.join(train_path, filename)
    if full_path.endswith(('.jpg', '.png', '.jpeg')):
        padded_image = resize_and_pad(full_path)
```

```python
for filename in os.listdir(test_path):
    full_path = os.path.join(test_path, filename)
    if full_path.endswith(('.jpg', '.png', '.jpeg')):
        padded_image = resize_and_pad(full_path)
```

```python
for filename in os.listdir(valid_path):
    full_path = os.path.join(valid_path, filename)
    if full_path.endswith(('.jpg', '.png', '.jpeg')):
        padded_image = resize_and_pad(full_path)
```

```python
print("Train images:")
check_image_shapes(train_path)
print("\nTest images:")
check_image_shapes(test_path)
print("\nvalid images:")
check_image_shapes(test_path)
```
